```python
import pandas as pd

reviews = pd.read_csv("scheme_reviews.csv")
reviews.head()
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

reviews["Score"] = reviews["Review_Text"].apply(get_sentiment)
reviews.head()
```

```python
def label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

reviews["Sentiment"] = reviews["Score"].apply(label)
reviews.head()
```

```python
import matplotlib.pyplot as plt

reviews["Sentiment"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", figsize=(6,6)
)
plt.title("Public Sentiment on Government Schemes")
plt.ylabel("")
plt.show()
```

Insight:
The pie chart shows that most public sentiment is positive,
indicating general satisfaction with government welfare schemes.
A smaller portion of users expressed negative or neutral opinions.


Conclusion:
Sentiment analysis using NLP techniques reveals that the majority
of public reviews are positive. This suggests that government schemes
are generally well received by the public.
