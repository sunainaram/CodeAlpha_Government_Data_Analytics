```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("government_schemes.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="State", y="Beneficiaries", data=df)
plt.xticks(rotation=45)
plt.title("Beneficiaries by State")
plt.xlabel("State")
plt.ylabel("Number of Beneficiaries")
plt.show()
```

Insight:
This bar chart shows the distribution of beneficiaries across different states.
Uttar Pradesh and Maharashtra have the highest number of beneficiaries, indicating
a wider reach of government welfare schemes in these regions.


```python
plt.figure(figsize=(8,5))
sns.barplot(x="Scheme_Name", y="Beneficiaries", data=df)
plt.title("Beneficiaries by Scheme")
plt.xlabel("Scheme")
plt.ylabel("Beneficiaries")
plt.show()
```

Insight:
This graph highlights that schemes like MNREGA and PMJDY have higher beneficiary
counts compared to others. These schemes play a significant role in employment
generation and financial inclusion.


```python
scheme_budget = df.groupby("Scheme_Name")["Budget_Crore"].sum()

plt.figure(figsize=(7,7))
scheme_budget.plot(kind='pie', autopct='%1.1f%%')
plt.title("Budget Distribution by Scheme")
plt.ylabel("")
plt.show()
```

Insight:
The pie chart represents the budget allocation among different government schemes.
It shows that a major portion of the budget is allocated to housing and employment
related schemes.


```python
pivot = df.pivot_table(values="Beneficiaries",
                       index="State",
                       columns="Scheme_Name",
                       aggfunc="sum")

plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, cmap="Blues")
plt.title("State vs Scheme Heatmap")
plt.show()
```

Insight:
The heatmap visualizes the relationship between states and schemes.
Darker shades indicate higher beneficiary counts, showing that MNREGA
has strong implementation across multiple states.


```python
plt.figure(figsize=(8,5))
sns.lineplot(x="Year", y="Beneficiaries", data=df)
plt.title("Beneficiary Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Beneficiaries")
plt.show()
```

Insight:
The line chart shows that the number of beneficiaries increased from 2020 to 2021
and slightly declined in 2022. This indicates changing trends in the
implementation of welfare schemes over time.


Conclusion:
This visualization project presents multiple graphical insights
into Indian government welfare schemes. Bar charts, pie charts,
line charts and heatmaps were used to understand beneficiary
distribution and budget allocation across states and schemes.
These visualizations help in better decision making and policy analysis.
