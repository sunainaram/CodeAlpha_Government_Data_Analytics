import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("government_schemes.csv")
df.head()


```python
df.info()
df.describe()
```

```python
df.sort_values(by='Beneficiaries', ascending=False).head()
```

```python
state_data = df.groupby('State')['Beneficiaries'].sum()
state_data
```

```python
import matplotlib.pyplot as plt

state_data.plot(kind='bar')
plt.title("State-wise Total Beneficiaries")
plt.xlabel("State")
plt.ylabel("Total Beneficiaries")
plt.show()
```

Insight:
The bar chart shows that Uttar Pradesh and Maharashtra have the highest number
of beneficiaries. This indicates that government schemes have wider reach
in these states compared to others.


```python
df.head()
df.info()
df.describe()
```

```python
scheme_benef = df.groupby("Scheme_Name")["Beneficiaries"].sum()

plt.figure(figsize=(8,5))
scheme_benef.plot(kind='bar')
plt.title("Scheme-wise Beneficiaries")
plt.ylabel("Beneficiaries")
plt.show()
```

Insight:
This graph indicates that MNREGA and PMJDY have the highest beneficiary counts.
These schemes play a major role in employment generation and financial inclusion.


```python
sector_benef = df.groupby("Sector")["Beneficiaries"].sum()

plt.figure(figsize=(7,7))
sector_benef.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sector-wise Distribution")
plt.show()
```

Insight:
The pie chart shows that the employment and banking sectors cover a large
portion of beneficiaries. This suggests that government policies are
focused on economic and financial development.


```python
year_trend = df.groupby("Year")["Beneficiaries"].sum()

plt.plot(year_trend)
plt.title("Year-wise Trend")
plt.xlabel("Year")
plt.ylabel("Beneficiaries")
plt.show()
```

Insight:
The line chart shows that the number of beneficiaries increased from 2020 to 2021
and slightly decreased in 2022. This reflects changing trends in scheme
implementation over time.


Conclusion:
The exploratory data analysis reveals that Indian government welfare schemes
have significant impact across multiple states and sectors. States like Uttar
Pradesh show higher beneficiary counts, while schemes such as MNREGA and PMJDY
play a crucial role in improving employment and financial inclusion.
